```python
df=pd.DataFrame()
for root, dirs, files in os.walk(os.getcwd()):
    for file in files:
        if file==("listings.csv"):
            #print(os.path.join(root, file))
            dflistings=pd.read_csv(os.path.join(root, file),encoding = "ISO-8859-1") 
            #print(root)
            RMS=os.path.basename(os.path.normpath(root))
            RMS=RMS.replace(',','')
            #print(RMS)
            Month=RMS.split(' ')[1]
            Year=RMS.split(' ')[2]
            #print(Month)
            #print(Year)
            dflistings['Month']=Month
            dflistings['Year']=Year
            df=pd.concat([df,dflistings])
```

```python
# https://datascience.stackexchange.com/questions/49553/combining-latitude-longitude-position-into-single-feature
from math import radians, cos, sin, asin, sqrt

def single_pt_haversine(lat, lng, degrees=True):
    """
    'Single-point' Haversine: Calculates the great circle distance
    between a point on Earth and the (0, 0) lat-long coordinate
    """
    r = 6371 # Earth's radius (km). Have r = 3956 if you want miles

    # Convert decimal degrees to radians
    if degrees:
        lat, lng = map(radians, [lat, lng])

    # 'Single-point' Haversine formula
    a = sin(lat/2)**2 + cos(lat) * sin(lng/2)**2
    d = 2 * r * asin(sqrt(a)) 

    return d

df['harvesine_distance'] = [single_pt_haversine(latitude, longitude) for latitude, longitude in zip(df['latitude'], df['longitude'])]
```
